```python
#translation

import cv2
import numpy as np
img = cv2.imread('sakura.jpg')
num_rows, num_cols = img.shape[:2]
print(img.shape, img.dtype)

#image translation using translation_matrix:
#[1,0, 70] #水平移動70
#[0,1,110] #垂直移動110
#[0,0,  1]
translation_matrix = np.float32([ [1,0,70], [0,1,110] ])
img_translation1 = cv2.warpAffine(img, translation_matrix, (num_cols+70, num_rows+110), cv2.INTER_LINEAR)

#image translation using ROI
img_translation2 = np.zeros((num_rows+110, num_cols+70, 3), dtype=np.uint8) 
print(img_translation2.shape)
#print(img_translation2[110:num_rows+110 , 70:num_cols+70, :])
img_translation2[110:num_rows+110 , 70:num_cols+70, :] = img

cv2.imshow('Translation1', img_translation1)
cv2.imshow('Translation2', img_translation2)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#rotation: 旋轉30度，轉動之後影像會超出範圍，要縮小0.7
#RotationMatrix2D
#[cos -sin tx]
#[sin cos  ty]
#[0   0    1 ]

import cv2
import numpy as np
img = cv2.imread('sakura.jpg')
num_rows, num_cols = img.shape[:2]
rotation_matrix = cv2.getRotationMatrix2D((num_cols/2, num_rows/2), -30, 0.7) # (num_cols/2, num_rows/2) 以中心點旋轉
rotation_matrix = cv2.getRotationMatrix2D((100, 100), -30, 0.7) # (num_cols/2, num_rows/2) 以中心點旋轉
print(rotation_matrix)
img_rotation = cv2.warpAffine(img, rotation_matrix, (num_cols, num_rows))
cv2.imshow('Rotation', img_rotation)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#scaling
#http://paulbourke.net/miscellaneous/interpolation/

import cv2
img = cv2.imread('sakura.jpg')
img_scaled = cv2.resize(img,None,fx=1.2, fy=1.2, interpolation = cv2.INTER_LINEAR)
cv2.imshow('Scaling - Linear Interpolation', img_scaled)
img_scaled = cv2.resize(img,None,fx=1.2, fy=1.2, interpolation = cv2.INTER_CUBIC)
cv2.imshow('Scaling - Cubic Interpolation', img_scaled)
img_scaled = cv2.resize(img,(600, 400), interpolation = cv2.INTER_AREA)
cv2.imshow('Scaling - Skewed Size', img_scaled)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#perspective transformation

import cv2
import numpy as np
img = cv2.imread('lena.png')
rows, cols = img.shape[:2]
src_points = np.float32([[0,0],             [cols-1,0], [0,rows-1],        [cols-1,rows-1]])
dst_points = np.float32([[0,int(0.4*rows)], [cols-1,0], [0,int(0.6*rows)], [cols-1,rows-1]])
#cv2.findHomography() can accept more than 4 pairs of corresponding points
projective_matrix = cv2.getPerspectiveTransform(src_points, dst_points)
print(projective_matrix)
img_output = cv2.warpPerspective(img, projective_matrix, (cols,rows), cv2.INTER_AREA, cv2.BORDER_CONSTANT, 0)

img_scaled = cv2.resize(img_output,(1920, 1080), interpolation = cv2.INTER_AREA)
cv2.imshow('Scaled', img_scaled)
cv2.imshow('Input', img)
cv2.imshow('Output', img_output)
ret,mask = cv2.threshold(img_output,1,255,cv2.THRESH_BINARY)
cv2.imshow('Mask', mask)
cv2.waitKey()
cv2.destroyAllWindows()
```

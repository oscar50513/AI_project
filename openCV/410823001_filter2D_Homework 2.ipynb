## Homework 2 (Morphological Filter)
1. Input image from gear_tooth.png.
2. The objective is to detect missing teeth using a sequence of morphological filters.
3. Optional: Design a set of structure elements: <b>gear_body</b>, <b>sampling_ring_spacer</b>, <b>sampling_ring_width</b>, <b>tip_spacing</b>, <b>defect_cue</b>
4. Apply morphological filters using the structure elements to generate the figures (e), (f), (g), (h)
<br>&nbsp;(e) opening it with structure element <b>gear_body</b> to get rid of the teeth, dilating that with structure element <b>sampling_ring_spacer</b> to bring it out to the base of the teeth, dilating that with the structure element <b>sampling_ring_width</b> to bring the next image out to the tip of the teeth, and subtracting the last two results to get a ring that just fits over the teeth.
<br>&nbsp;(f) AND it with the original image to produce an image of just the teeth.
<br>&nbsp;(g) dilating the teeth image with the structure element <b>tip_spacing</b> produces the solid ring image which has spaces in the solid ring wherever there are defects in the teeth.
<br>&nbsp;(h) subtracting it from the sampling ring leaves only the defects, which are dilated by structure element <b>defect_cue</b>.
5. Mark the defected gears using <b>red rectangles</b> on the output images.
6. Write a simple report in a separate cell.
7. Upload your Jupyter code file (*.ipynb)


```python
#gear tooth inspection using morphological filter

import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('gear_tooth.png', cv2.IMREAD_GRAYSCALE)
retval, img_a = cv2.threshold(img, 225, 255, cv2.THRESH_BINARY)

# 找到小圈圈 hole_ring
kernel_dot = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
kernel_hole = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(30,30))
kernel_hole_e = cv2.erode(kernel_hole, kernel_dot, iterations=1)
kernel_hole_ring = kernel_hole - kernel_hole_e

# 找到大圈圈 hole_mask
kernel_dot2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
kernel_hole2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(79,79))
kernel_hole_e2 = cv2.erode(kernel_hole2, kernel_dot2, iterations=1)
gear_body = kernel_hole2 - kernel_hole_e2


img_b = cv2.erode(img_a, kernel_hole_ring, iterations=1)       # erode 找到缺口的位置
img_c = cv2.dilate(img_b, kernel_hole, iterations=1)           # 在位置上 dilate 擴張成 kernel_hole 大小
img_d = cv2.bitwise_or(img_a, img_c)                           # 填滿孔洞大小 做 OR

img_e_dot = cv2.erode(img_d, gear_body, iterations=1, borderType=cv2.BORDER_CONSTANT,borderValue=0)  # 找到齒輪的中心點位置
sampling_ring_spacer = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(81,82)) # 內部圈圈
img_e1 = cv2.dilate(img_e_dot, sampling_ring_spacer, iterations=1, borderType = cv2.BORDER_CONSTANT,borderValue=255) 
sampling_ring_width = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(87,87)) # 外部圈圈
img_e2 = cv2.dilate(img_e_dot, sampling_ring_width, iterations=1, borderType = cv2.BORDER_CONSTANT,borderValue=0)
img_e = img_e2-img_e1


img_f = cv2.bitwise_and(img, img_e)
retval, img_f = cv2.threshold(img_f, 100, 255, cv2.THRESH_BINARY)

tip_spacing = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
img_g = cv2.dilate(img_f, tip_spacing, iterations=1)

defect_cue = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
img_h_error = img_e-img_g
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(2,2))
img_h_dot = cv2.morphologyEx(img_h_error, cv2.MORPH_OPEN, kernel) #先收縮再擴張
img_h_error = cv2.dilate(img_h_dot, defect_cue, iterations=2)
img_h = cv2.bitwise_or(img_h_error, img_g)  

# 建立矩形外框
rect_body = cv2.getStructuringElement(cv2.MORPH_RECT,(15,15)) # 內部圈圈
img_r1 = cv2.dilate(img_h_dot, rect_body, iterations=1, borderType = cv2.BORDER_CONSTANT,borderValue=255)
rect_body2 = cv2.getStructuringElement(cv2.MORPH_RECT,(17,17)) # 內部圈圈
img_r2 = cv2.dilate(img_h_dot, rect_body2, iterations=1, borderType = cv2.BORDER_CONSTANT,borderValue=0)
img_r = img_r2-img_r1
h,w = img_r.shape
img_red = np.zeros((h,w,3),dtype=np.uint8)
img_red[:,:,2] = img_r 
Original = cv2.imread('gear_tooth.png')
Result = cv2.addWeighted(Original,0.5,img_red,0.5,0)
hsv = cv2.cvtColor(Result, cv2.COLOR_BGR2HSV)
s = hsv[:,:,1]
mask = cv2.inRange(s,0,50)
Result = np.copy(Original)
Result[mask==0]=[0,0,255]
cv2.imshow('Original', Original)
cv2.imshow('Result', Result)
cv2.waitKey()
cv2.destroyAllWindows()

titles = ['image(A)','image(B)','image(C)','image(D)','image(E)','image(F)','image(G)','image(H)','red rectangles','Result']
img_red = cv2.cvtColor(img_red,cv2.COLOR_BGR2RGB)
Result = cv2.cvtColor(Result,cv2.COLOR_BGR2RGB)
imgs = [img_a,img_b,img_c,img_d,img_e,img_f,img_g,img_h,img_red,Result]

plt.figure(figsize=(6,10))
for i in range(10):
    plt.subplot(5,2,i+1)
    plt.title(titles[i])
    plt.imshow(imgs[i], cmap = 'gray')
    plt.axis('off')

plt.tight_layout()
plt.show()
print()

#gear_body, sampling_ring_spacer, sampling_ring_width, tip_spacing, defect_cue,img_h_dot
titles = ['effect of image (E)','dilate img_e_dot','with sampling_ring_spacer','= img_e1','dilate img_e_dot','with sampling_ring_width','= img_e2','img_e2','- img_e1','img_e']
imgs = [img_e_dot,sampling_ring_spacer,img_e1,img_e_dot,sampling_ring_width,img_e2,img_e2,img_e1,img_e,Result]
plt.figure(figsize=(8,10))
plt.suptitle('effect of image (E)',fontsize=20,color='red')
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.title(titles[i], fontsize=10)
    plt.imshow(imgs[i], cmap = 'gray')
    plt.axis('off')
plt.tight_layout()
plt.show()
print()

plt.figure(figsize=(8,10))
plt.suptitle('effect of image (F)',fontsize=20,color='red')
plt.subplot(3,3,1)
plt.title('image(A)', fontsize=10)
plt.imshow(img_a, cmap = 'gray')
plt.axis('off')
plt.subplot(3,3,2)
plt.title('AND image(E)', fontsize=10)
plt.imshow(img_e, cmap = 'gray')
plt.axis('off')
plt.subplot(3,3,3)
plt.title('= image (F)', fontsize=10)
plt.imshow(img_f, cmap = 'gray')
plt.axis('off')
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,10))
plt.suptitle('effect of image (G)',fontsize=20,color='red')
plt.subplot(3,3,1)
plt.title('dilate image (F)', fontsize=10)
plt.imshow(img_f, cmap = 'gray')
plt.axis('off')
plt.subplot(3,3,2)
plt.title('with tip_spacing', fontsize=10)
plt.imshow(tip_spacing, cmap = 'gray')
plt.axis('off')
plt.subplot(3,3,3)
plt.title('= image (G)', fontsize=10)
plt.imshow(img_g, cmap = 'gray')
plt.axis('off')
plt.tight_layout()
plt.show()
print()
print()
plt.figure(figsize=(8,10))
plt.suptitle('effect of image (H)',fontsize=20,color='red')
plt.subplot(3,3,1)
plt.title('image(E)', fontsize=10)
plt.imshow(img_e, cmap = 'gray')
plt.axis('off')
plt.subplot(3,3,2)
plt.title(' - image(G)', fontsize=10)
plt.imshow(img_g, cmap = 'gray')
plt.axis('off')
plt.subplot(3,3,3)
plt.title('= img_h_dot', fontsize=10)
plt.imshow(img_h_dot, cmap = 'gray')
plt.axis('off')
plt.subplot(3,3,4)
plt.title('dilate img_h_dot', fontsize=10)
plt.imshow(img_h_dot, cmap = 'gray')
plt.axis('off')
plt.subplot(3,3,5)
plt.title('with defect_cue', fontsize=10)
plt.imshow(defect_cue, cmap = 'gray')
plt.axis('off')
plt.subplot(3,3,6)
plt.title('= img_h_error', fontsize=10)
plt.imshow(img_h_error, cmap = 'gray')
plt.axis('off')
plt.subplot(3,3,7)
plt.title('img_h_error', fontsize=10)
plt.imshow(img_h_error, cmap = 'gray')
plt.axis('off')
plt.subplot(3,3,8)
plt.title('OR image(G)', fontsize=10)
plt.imshow(img_g, cmap = 'gray')
plt.axis('off')
plt.subplot(3,3,9)
plt.title('image(H)', fontsize=10)
plt.imshow(img_h, cmap = 'gray')
plt.axis('off')
plt.tight_layout()
plt.show()

print()
plt.figure(figsize=(8,10))
plt.suptitle('effect of the Red Rectangles',fontsize=20,color='red')
plt.subplot(3,3,1)
plt.title('dilate img_h_dot', fontsize=10)
plt.imshow(img_h_dot, cmap = 'gray')
plt.axis('off')
plt.subplot(3,3,2)
plt.title('with rect_body', fontsize=10)
plt.imshow(rect_body, cmap = 'gray')
plt.axis('off')
plt.subplot(3,3,3)
plt.title('= img_r1', fontsize=10)
plt.imshow(img_r1, cmap = 'gray')
plt.axis('off')
plt.subplot(3,3,4)
plt.title('dilate img_h_dot', fontsize=10)
plt.imshow(img_h_dot, cmap = 'gray')
plt.axis('off')
plt.subplot(3,3,5)
plt.title('with rect_body2', fontsize=10)
plt.imshow(rect_body2, cmap = 'gray')
plt.axis('off')
plt.subplot(3,3,6)
plt.title('= img_r2', fontsize=10)
plt.imshow(img_r2, cmap = 'gray')
plt.axis('off')
plt.subplot(3,3,7)
plt.title('img_r2', fontsize=10)
plt.imshow(img_r2, cmap = 'gray')
plt.axis('off')
plt.subplot(3,3,8)
plt.title('- img_r1', fontsize=10)
plt.imshow(img_r1, cmap = 'gray')
plt.axis('off')
plt.subplot(3,3,9)
plt.title('img_r', fontsize=10)
plt.imshow(img_r, cmap = 'gray')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
#write video file

import cv2
import numpy as np

cap = cv2.VideoCapture('WiiPlay.mp4')
if not cap.isOpened():
    raise IOError("Cannot open the video file")

def Thresholding(pos):
    pass


# 設定開始和結束的影格
# Set the start and end frames
start_time = 20*60+41
end_time = 20*60+50+1000

fps = 30
frame_seq = start_time * fps # 計算開始的影格 to calculate the starting frame
out_size = (640, 360)


# fourcc = cv2.VideoWriter_fourcc(*'mp4v')
# out = cv2.VideoWriter('40621204_output.mp4', fourcc, 30.0, out_size)

cap.set(cv2.CAP_PROP_POS_FRAMES , start_time*fps);


ret, frame = cap.read()
first_frame = cv2.resize(frame, out_size, 0, 0, interpolation=cv2.INTER_AREA)


# 創建 Hue 色彩頻道的上下限 createTrackbar
# Create the upper and lower limits of the Hue color channel createTrackbar
cv2.namedWindow('Show_HSV_Threshold')

cv2.createTrackbar('H_upper', 'Show_HSV_Threshold', 179, 179, Thresholding) 
cv2.createTrackbar('H_lower', 'Show_HSV_Threshold', 0, 179, Thresholding) 


count = start_time * fps
while True:
    ret, frame = cap.read()
    
    if ret == False  or count == end_time* fps:
        break
    else:
        count += 1 
    
    cur_frame = cv2.resize(frame, out_size, 0, 0, interpolation=cv2.INTER_AREA)
    cv2.imshow('Input', cur_frame)

    
    # 將畫面 RGB 改為 HSV
    # Change picture RGB to HSV
    cur_frame_HVS = cv2.cvtColor(cur_frame, cv2.COLOR_BGR2HSV)
    
    # 取得 Trackbar 即時設定的 Hue 上下限數值
    # Get the upper and lower limit values of Hue set by Trackbar in real time
    H_upper = cv2.getTrackbarPos('H_upper', 'Show_HSV_Threshold')
    H_lower = cv2.getTrackbarPos('H_lower', 'Show_HSV_Threshold')

    # 將上下限數值放進numpy array 中
    # Put the upper and lower limit values into the numpy array
    upper = np.array([H_upper])
    lower = np.array([H_lower])
    
    # 創建 Hue 頻道的遮罩
    # Create a mask for the Hue channel
    mask = cv2.inRange(cur_frame_HVS[:,:,0],lower,upper)

    cv2.imshow('mask', mask)
    cv2.imshow('HVS_cur_frame', cur_frame_HVS)
    cv2.imshow('cur_frame', cur_frame)
    
    c = cv2.waitKey(10)
    if c == 27:
        break
        
cap.release()
# out.release()
cv2.destroyAllWindows()
```

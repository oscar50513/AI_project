```python
# Keras中的模型應用(不使用遷移式學習) (以MobileNetV3Small為例)
import tensorflow.keras.applications as keras_model
from tensorflow.keras.layers import Input
model = keras_model.MobileNetV3Small(include_top=True,   # 包含頂部的全連接層
                                     weights=None,       # 不使用ImageNet權重
                                     input_tensor=Input(shape=(224, 224, 3)), # 張量形狀設為(224, 224, 3)
                                     input_shape=None, classes=2) # input_shape : 可選的形狀元組
model.summary()
```

```python
# Keras中的模型應用(使用遷移式學習) (以InceptionV3為例)
import tensorflow.keras.applications as keras_model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
base_model = keras_model.InceptionV3(include_top=False, weights="imagenet",
                                input_shape=(224, 224, 3), pooling=None, classes=2)
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation="relu")(x)
x = Dense(512, activation="relu")(x)
output = Dense(2, activation="softmax")(x)
model = Model(inputs=base_model.input, outputs=output)
model.summary()
```

```python
from keras.preprocessing.image import ImageDataGenerator
import cv2
datagen = ImageDataGenerator(
    zca_whitening=False,
    rotation_range=180,
    #shear_range=0.2,
    #zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='constant',
    cval=255)
```

```python
# 

import cv2
from matplotlib import pyplot as plt
img = cv2.imread(r"C:\Users\Landis\Desktop\yolov7-main\coffee bean\val data\test01.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#(sample, channels, height, width)
img = img.reshape((1,) + img.shape)
i = 0
for batch in datagen.flow(img, batch_size=10,
                          save_to_dir=r'C:\Users\Landis\Desktop\yolov7-main\coffee bean\preview', save_prefix='good_bean', save_format='jpeg'):
    plt.subplot(3,3,1 + i)
    plt.axis("off")
    augImage = batch[0]
    augImage = augImage.astype('float32')
    augImage /= 255
    plt.imshow(augImage)
    i += 1
    if i > 8:
        break
```
